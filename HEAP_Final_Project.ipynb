```python
df = df0.merge(df1, left_on='X', right_on='X')
df = df.merge(df2, left_on='X', right_on='X')
df = df.merge(df3, left_on='X', right_on='X')
df = df.merge(df4, left_on='X', right_on='X')
df = df.merge(df5, left_on='X', right_on='X')
df = df.merge(df6, left_on='X', right_on='X')
df = df.merge(df7, left_on='X', right_on='X')
df = df.merge(df8, left_on='X', right_on='X')
df = df.merge(df9, left_on='X', right_on='X')
df = df.merge(df10, left_on='X', right_on='X')
df.head(5)
```

```python
def area_under(x):
    area = 0
    for i in range(1799):
        box = (df['X'].iloc[i+1] - df['X'].iloc[i])*(df[x].iloc[i])
        area += box   
    return area
```
